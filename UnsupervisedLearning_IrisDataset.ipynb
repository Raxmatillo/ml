```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.datasets import load_iris

iris = load_iris()
```

```python
iris.keys()
```

```python
iris.data[:5]
```

```python
from sklearn.cluster import KMeans

X = iris.data
kmeans = KMeans(n_clusters=3, init='k-means++', random_state=1)
kmeans.fit(X)
```

```python
labels = kmeans.labels_

plt.scatter(X[:, 0], X[:, 2], c=labels, s=50, cmap='viridis')
plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')
plt.show()
```

```python
# bestk k
inertia = []

for k in range(1, 10):
    kmeans=KMeans(n_clusters=k, init='k-means++', random_state=1)
    kmeans.fit(X)
    inertia.append(kmeans.inertia_)

plt.plot(range(1, 10), inertia)
plt.title('Elbow method')
plt.xlabel('Clusters k')
plt.ylabel('Inertia')
plt.show()
```

```python
# Silhouette score
from sklearn.metrics import silhouette_score
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

silhouette_scores = []
for k in range(2, 10):
    kmeans=KMeans(n_clusters=k, init='k-means++', random_state=1)
    kmeans.fit(X_scaled)
    labels = kmeans.labels_
    score = silhouette_score(X_scaled, labels)
    silhouette_scores.append(score)

plt.plot(range(2, 10), silhouette_scores, color='green')
plt.title('Silhouette score method')
plt.xlabel('Clusters k')
plt.ylabel('Silhouette')
plt.show()
```

### DBSCAN

```python
from sklearn.cluster import DBSCAN
```

```python
db = DBSCAN(eps=0.7, min_samples=5)
db.fit(X)
```

```python
core_sample_mask = np.zeros_like(db.labels_, dtype=bool)
core_sample_mask[db.core_sample_indices_] = True
labels = db.labels_
```

```python
n_clusters_ = len(set(labels)) - (1 if -1 in labels else 0)
```

```python
unique_labels = set(labels)
colors = ['y', 'b', 'g', 'r']
for k, col in zip(unique_labels, colors):
    if k==-1:
        col = 'k'
    class_member_mask = (labels == k)

    xy = X[class_member_mask & core_sample_mask]
    plt.plot(xy[:, 0], xy[:, 2], 'o', markerfacecolor=col, markeredgecolor='k', markersize=6)
    xy = X[class_member_mask & ~core_sample_mask]
    plt.plot(xy[:, 0], xy[:, 2], 'o', markerfacecolor=col, markeredgecolor='k', markersize=6)

plt.title('number of clusters: %d' % n_clusters_)
plt.show()
```
