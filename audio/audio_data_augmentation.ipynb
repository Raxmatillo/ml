```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import IPython.display as ipd 
import librosa
import librosa.display
```

```python
audio_path = 'clips/common_voice_uz_41832983.mp3'
data, sr = librosa.load(audio_path)
```

```python
ipd.Audio(audio_path)
```

```python
len(data)
```

```python
plt.figure(figsize=(12, 4))
plt.figure()
librosa.display.waveshow(data, sr=sr)
```

```python
def add_white_noise(signal, noise_precentage_factor):
    noise = np.random.normal(0, signal.std(), signal.size)
    augmented_signal = signal + noise * noise_precentage_factor
    return augmented_signal
```

```python
augmented_signal = add_white_noise(data, 0.3)
```

```python
import soundfile as sf

sf.write('augmented_audio_white_noise.wav', augmented_signal, sr)
```

```python
ipd.Audio('augmented_audio_white_noise.wav')
```

```python
plt.figure(figsize=(12, 4))
plt.figure()
librosa.display.waveshow(augmented_signal, sr=sr)
```

```python
def time_stretch(signal, time_stretch_rate):
    return librosa.effects.time_stretch(signal, rate=time_stretch_rate)
```

```python
augmented_signal = time_stretch(data, 1.1)
sf.write('augmented_audio_time_stretch.wav', augmented_signal, sr)
```

```python
ipd.Audio('augmented_audio_time_stretch.wav')
```

```python
def pitch_scale(signal, sr = 22050, num_semitones = 2):
    return librosa.effects.pitch_shift(y=signal, sr=sr, n_steps=num_semitones)
```

```python
augmented_signal = pitch_scale(data, num_semitones=4)
sf.write('augmented_audio_pitch_scale.wav', augmented_signal, sr)
```

```python

ipd.Audio('augmented_audio_pitch_scale.wav')
```
